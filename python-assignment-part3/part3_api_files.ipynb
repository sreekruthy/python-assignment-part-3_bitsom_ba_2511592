Part B - Read:

```python
# read file back and print each line numbered, stripping the trailing newline character (\n) from each line
with open("python_notes.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()   # read all lines into list

# print numbered lines
for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")   # strip() removes '\n'

print("\nTotal lines:", len(lines)) # count and print total no of lines in file


# ask the user for a keyword
keyword = input("Enter keyword to search: ").lower()

found = False

# print all lines that contain that keyword (case-insensitive)
for line in lines:
    if keyword in line.lower():
        print(line.strip())
        found = True

# if no match is found, print a helpful message
if not found:
    print("No matching lines found.")
```

Step 1 - Fetch and Display Products:

```python
import requests

# API endpoint to fetch 20 products
url = "https://dummyjson.com/products?limit=20"

response = requests.get(url) # GET request to API
data = response.json() # parse JSON response

products = data["products"] # extract list of products from response

print("ID | Title | Category | Price | Rating")
print("-"*60)

# print formatted table
for p in products:
    print(f"{p['id']:<2} | {p['title']:<45} | {p['category']:<10} | ${p['price']:<7} | {p['rating']}")
```

Step 2 - Filter and Sort:

```python
# from the 20 fetched products, filter only those with a rating ≥ 4.5
filtered = [p for p in products if p["rating"] >= 4.5]

# sort the filtered list by price in descending order
filtered.sort(key=lambda x: x["price"], reverse=True)

# print the filtered and sorted list
print("Filtered & Sorted Products:\n")
for p in filtered:
    print(f"{p['title']} - ${p['price']} - Rating: {p['rating']}")
```

Step 3 - Search by Category:

```python
# API endpoint for laptops category
url = "https://dummyjson.com/products/category/laptops"

response = requests.get(url) # GET request to fetch all products in the laptops category
data = response.json() # parse JSON response

# print the name and price of each laptop found
print("\nLaptops:")
for p in data["products"]:
    print(f"{p['title']} - ${p['price']}")
```

Step 4 - POST Request (Simulated):

```python
# API endpoint to add products
url = "https://dummyjson.com/products/add"

# JSON body
payload = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

response = requests.post(url, json=payload) # POST request with JSON data

print("\nPOST Response:")
print(response.json()) # convert response to JSON and print
```

#Task 3 — Exception Handling

Part A - Guarded Calculator:

```python
def safe_divide(a, b):
    try:
        return a / b

    except ZeroDivisionError:
        return "Error: Cannot divide by zero"

    except TypeError:
        return "Error: Invalid input types"


# Testing
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
```

Part B - Guarded File Reader:

```python
def read_file_safe(filename):
    try:
        with open(filename, "r") as file: #open given file
            return file.read() # read the given file, then return its full content as a string

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")

    finally:
        print("File operation attempt complete.")


# Testing
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))
```

Part C - Robust API Calls:

```python
import requests

# GET requests to fetch 20 products

url = "https://dummyjson.com/products?limit=20" # API endpoint

# requests.get() call is wrapped in a try-except block
try:
    response = requests.get(url, timeout=5)
    data = response.json()
    print(data)

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

```python
# GET requests for laptops category

url = "https://dummyjson.com/products/category/laptops" # API endpoint

# requests.get() call is wrapped in a try-except block
try:
    response = requests.get(url, timeout=5)
    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

```python
# POST request

url = "https://dummyjson.com/products/add"

payload = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

# requests.post() call is wrapped in a try-except block
try:
    response = requests.post(url, json=payload, timeout=5)
    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

Part D - Input Validation Loop:

```python
import requests

while True:
    user_input = input("\nEnter product ID to look up (1–100), or 'quit' to exit: ") # asks the user

    if user_input.lower() == "quit": # continues until the user types quit
        break

    # validate that the input is an integer in the range 1–100; if not, print a warning
    if not user_input.isdigit():
        print("Do not make an API call with invalid input. Enter a number.")
        continue

    pid = int(user_input) # ask again

    if pid < 1 or pid > 100:
        print("Enter ID between 1 and 100.")
        continue

    # API call
    url = f"https://dummyjson.com/products/{pid}"
    response = requests.get(url) # GET request

    # checks response.status_code
    if response.status_code == 404:
        print("Product not found.")

    elif response.status_code == 200:
        data = response.json()
        print(f"{data['title']} - ${data['price']}")
```
